```python
import snowflake.connector

# Connect to Snowflake
conn = snowflake.connector.connect(
    user='cds006',
    password='@Triton123',
    account='qla31786.east-us-2.azure'
)

# Create a cursor object to execute SQL commands
cs = conn.cursor()

# Step 1: Create a Virtual Warehouse
cs.execute("CREATE WAREHOUSE IF NOT EXISTS my_first_warehouse")
print("Virtual Warehouse created successfully.")

# Set the active warehouse (optional if it's already active)
cs.execute("USE WAREHOUSE my_first_warehouse")

# Step 2: Create a Database
cs.execute("CREATE DATABASE IF NOT EXISTS testdb")
print("Database created successfully.")

# Set the active database (optional if it's already active)
cs.execute("USE DATABASE testdb")

# Step 3: Create a Schema
cs.execute("CREATE SCHEMA IF NOT EXISTS testschema")
print("Schema created successfully.")

# Set the active schema (optional if it's already active)
cs.execute("USE SCHEMA testschema")

# Step 4: Create a Table and Insert Data
cs.execute("""
    CREATE OR REPLACE TABLE test_table (
        col1 INTEGER,
        col2 STRING
    )
""")
print("Table created successfully.")

# Insert sample data into the table
cs.execute("""
    INSERT INTO test_table (col1, col2)
    VALUES
        (123, 'test string1'),
        (456, 'test string2')
""")
print("Data inserted into table successfully.")

# Step 5: Create a Stage
# Creating an internal stage to store files temporarily
cs.execute("CREATE STAGE IF NOT EXISTS test_table_stage")
print("Internal stage created successfully.")

# Load Data using PUT and COPY INTO Commands
# Put the files into the Snowflake stage
cs.execute("PUT file:///tmp/data/file* @test_table_stage")
print("Files placed into stage successfully.")

# Copy the data from the stage into the table
cs.execute("COPY INTO test_table FROM @test_table_stage")
print("Data copied into table successfully.")

# Step 6: Run a Query and View the Result Set
cs.execute("SELECT * FROM test_table")
# Fetch and print the first two rows
print(cs.fetchmany(2))

# Close the cursor and connection
cs.close()
conn.close()
```
